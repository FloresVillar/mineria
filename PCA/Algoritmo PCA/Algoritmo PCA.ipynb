El **Análisis de Componentes Principales (PCA)** es una técnica matemática de reducción de dimensionalidad que transforma un conjunto de variables correlacionadas en un nuevo conjunto de variables no correlacionadas llamadas **Componentes Principales (CP)**.

Matemáticamente, PCA busca las direcciones (vectores) donde los datos tienen la **máxima varianza**. Aquí tienes el desglose paso a paso:

---

#### 1. Estandarización de los datos
PCA es extremadamente sensible a las escalas (como vimos antes). El primer paso es centrar los datos restando la media ($\mu$) y, opcionalmente, dividiendo por la desviación estándar ($\sigma$).

Si $X$ es nuestra matriz de datos ($n$ muestras, $d$ dimensiones):
$$Z = \frac{X - \mu}{\sigma}$$
Esto asegura que todas las variables tengan media 0 y varianza 1.

---

#### 2. Matriz de Covarianza
Buscamos entender cómo cambian las variables unas respecto a otras. Calculamos la matriz de covarianza $\Sigma$ (o $C$):

$$\Sigma = \frac{1}{n-1} Z^T Z$$

*   Es una matriz cuadrada de $d \times d$.
*   Los elementos de la **diagonal** son las varianzas de cada variable.
*   Los elementos **fuera de la diagonal** son las covarianzas (qué tanto "se mueven juntas" dos variables).

---

#### 3. Cálculo de Autovalores y Autovectores
Este es el corazón de PCA. Resolvemos la ecuación característica de la matriz de covarianza para encontrar sus **autovectores** ($v$) y **autovalores** ($\lambda$):

$$\Sigma \mathbf{v} = \lambda \mathbf{v}$$

*   **Autovectores ($\mathbf{v}$):** Representan las **direcciones** de los nuevos ejes (Componentes Principales). Son ortogonales entre sí (forman un ángulo de 90°).
*   **Autovalores ($\lambda$):** Representan la **magnitud** de la varianza explicada por cada autovector.

---

#### 4. Selección de Componentes Principales
Ordenamos los autovalores de mayor a menor: $\lambda_1 \ge \lambda_2 \ge \dots \ge \lambda_d$.

*   El autovector correspondiente a $\lambda_1$ es el **Primer Componente Principal (PC1)**. Es la dirección donde los datos están más "estirados".
*   Para reducir dimensiones de $d$ a $k$, seleccionamos los primeros $k$ autovectores y formamos una matriz de proyección $W$.

**¿Cuánta información retenemos?**
La varianza explicada por el componente $i$ es:
$$\text{Varianza explicada} = \frac{\lambda_i}{\sum_{j=1}^d \lambda_j}$$

---

#### 5. Proyección al nuevo espacio
Finalmente, multiplicamos nuestros datos originales estandarizados ($Z$) por la matriz de autovectores seleccionados ($W$):

$$Y = Z \cdot W$$

Donde $Y$ es el nuevo dataset con dimensiones reducidas.

---

#### Resumen Visual de la Matemática
Imagina una nube de puntos con forma de elipse en 2D:
1.  **PC1:** Es el vector que atraviesa la parte más larga de la elipse (máxima varianza).
2.  **PC2:** Es el vector perpendicular a PC1 que captura el resto de la varianza.
3.  **Resultado:** Si decides que 1D es suficiente, "aplastas" (proyectas) todos los puntos sobre la línea del PC1, perdiendo la información del PC2 pero manteniendo la estructura principal de los datos.

#### Diferencia clave: PCA vs Regresión Lineal
*   La **Regresión Lineal** intenta minimizar la distancia *vertical* entre los puntos y la línea (error de predicción).
*   **PCA** intenta minimizar la distancia *perpendicular* (proyección) entre los puntos y el eje, para maximizar la varianza proyectada.

### Ejemplo númerico

Para explicar PCA en una clase de Machine Learning, lo mejor es usar un ejemplo con datos pequeños para que la aritmética sea visible.

Imagina que tenemos 3 estudiantes y medimos su **Estatura (cm)** y su **Peso (kg)**. Queremos reducir estas 2 variables a solo 1 variable (un "índice de tamaño corporal").

#### Nuestros Datos ($X$):
| Estudiante | Estatura ($x_1$) | Peso ($x_2$) |
| :--- | :--- | :--- |
| A | 170 | 60 |
| B | 180 | 80 |
| C | 160 | 70 |

---

#### Paso 1: Centrar los datos (Media = 0)
PCA necesita que los datos estén centrados en el origen $(0,0)$.
1.  Calculamos la media de cada columna:
    *   $\mu_{Estatura} = (170+180+160) / 3 = \mathbf{170}$
    *   $\mu_{Peso} = (60+80+70) / 3 = \mathbf{70}$
2.  Restamos la media a cada dato:

| Estudiante | $x_1 - \mu_1$ | $x_2 - \mu_2$ |
| :--- | :--- | :--- |
| A | 0 | -10 |
| B | 10 | 10 |
| C | -10 | 0 |

Matriz centrada **$Z$**:
$$Z = \begin{pmatrix} 0 & -10 \\ 10 & 10 \\ -10 & 0 \end{pmatrix}$$

---

#### Paso 2: Calcular la Matriz de Covarianza
Esta matriz nos dice cómo varían las variables juntas. La fórmula es $\text{Cov} = \frac{1}{n-1} Z^T Z$.
Para simplificar, calculemos $Z^T Z$:
$$Z^T Z = \begin{pmatrix} 0 & 10 & -10 \\ -10 & 10 & 0 \end{pmatrix} \begin{pmatrix} 0 & -10 \\ 10 & 10 \\ -10 & 0 \end{pmatrix} = \begin{pmatrix} 200 & 100 \\ 100 & 200 \end{pmatrix}$$
Dividimos por $n-1 = 2$:
$$\Sigma = \begin{pmatrix} 100 & 50 \\ 50 & 100 \end{pmatrix}$$

---

#### Paso 3: Calcular Autovalores ($\lambda$) y Autovectores ($v$)
Aquí es donde ocurre la "magia". Resolvemos $\det(\Sigma - \lambda I) = 0$:
$$\det \begin{pmatrix} 100-\lambda & 50 \\ 50 & 100-\lambda \end{pmatrix} = 0$$
$$(100-\lambda)^2 - 50^2 = 0 \implies (100-\lambda)^2 = 2500$$
Raíz cuadrada en ambos lados: $100-\lambda = \pm 50$.
*   $\lambda_1 = 150$ (Este componente explica la mayor varianza).
*   $\lambda_2 = 50$ (Este componente explica el resto).

Ahora buscamos el **Autovector** para $\lambda_1 = 150$ resolviendo $(\Sigma - 150I)v = 0$:
$$\begin{pmatrix} -50 & 50 \\ 50 & -50 \end{pmatrix} \begin{pmatrix} v_1 \\ v_2 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix} \implies -50v_1 + 50v_2 = 0 \implies v_1 = v_2$$
Normalizado, el autovector es: **$v_1 = [0.707, 0.707]$**.
*(Este vector apunta a 45 grados, indicando que Estatura y Peso pesan lo mismo en el nuevo índice).*

---

#### Paso 4: Varianza Explicada
¿Cuánta información conservamos si solo nos quedamos con el primer componente?
$$\frac{\lambda_1}{\lambda_1 + \lambda_2} = \frac{150}{150 + 50} = 0.75$$
**Conclusión:** Al reducir de 2D a 1D, conservamos el **75% de la varianza** (información) original.

---

#### Paso 5: Proyección (Reducción de dimensión)
Para obtener el nuevo valor (PC1) de cada estudiante, multiplicamos los datos centrados por el autovector $v_1$:
*   **Estudiante A:** $(0, -10) \cdot (0.707, 0.707) = \mathbf{-7.07}$
*   **Estudiante B:** $(10, 10) \cdot (0.707, 0.707) = \mathbf{14.14}$
*   **Estudiante C:** $(-10, 0) \cdot (0.707, 0.707) = \mathbf{-7.07}$

**Resultado final:** Hemos pasado de una tabla con 2 columnas (Estatura, Peso) a una con solo 1 columna (Tamaño Corporal).

---

#### Resumen para la clase:
1.  **¿Qué es el autovector?** Es la **dirección** de la línea que mejor atraviesa la nube de puntos.
2.  **¿Qué es el autovalor?** Es la **longitud** o fuerza de esa línea. Cuanto más grande es $\lambda$, más información captura ese eje.
3.  **¿Por qué es útil?** Si tienes 100 variables, PCA te permite ver que quizás con 3 "super-variables" (componentes) explicas el 95% de lo que pasa, simplificando mucho el modelo de Machine Learning.
