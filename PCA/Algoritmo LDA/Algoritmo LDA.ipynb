````python
El **LDA (Linear Discriminant Analysis)** o Análisis Discriminante Lineal es un algoritmo de **aprendizaje supervisado** utilizado para la reducción de dimensionalidad y la clasificación.

A diferencia del PCA (que busca las direcciones de máxima varianza sin importar las clases), el LDA busca las direcciones que **mejor separan a las clases**.

---

### 1. La Intuición de LDA
El objetivo de LDA es proyectar los datos en un espacio de menor dimensión maximizando dos cosas simultáneamente:
1.  **La distancia entre las medias** de las clases (Separación entre grupos).
2.  **La mínima varianza dentro de cada clase** (Compactación de cada grupo).

**Analogía:** Si PCA es como tomar una foto desde el ángulo donde los objetos están más dispersos, LDA es como tomar la foto desde el ángulo donde los grupos de objetos no se enciman unos con otros.

---

### 2. Las Matemáticas de LDA (Paso a Paso)

Para un conjunto de datos con $C$ clases:

#### Paso 1: Calcular los vectores de media ($\mathbf{m}_i$)
Calculamos el promedio de las características para cada clase de forma independiente.
$$\mathbf{m}_i = \frac{1}{n_i} \sum_{\mathbf{x} \in \text{Clase}_i} \mathbf{x}$$

#### Paso 2: Calcular las Matrices de Dispersión (Scatter Matrices)
Aquí es donde LDA se diferencia de PCA. Necesitamos dos matrices:

*   **Matriz de dispersión intra-clase ($S_W$ - Within-class):** Mide qué tan dispersos están los puntos dentro de su propia clase.
    $$S_W = \sum_{i=1}^C \sum_{\mathbf{x} \in \text{Clase}_i} (\mathbf{x} - \mathbf{m}_i)(\mathbf{x} - \mathbf{m}_i)^T$$
*   **Matriz de dispersión entre-clases ($S_B$ - Between-class):** Mide qué tan lejos están las medias de las clases entre sí.
    $$S_B = \sum_{i=1}^C n_i (\mathbf{m}_i - \mathbf{m})(\mathbf{m}_i - \mathbf{m})^T$$
    *(Donde $\mathbf{m}$ es la media de todo el dataset).*

#### Paso 3: Resolver el problema de autovalores
Buscamos una proyección $w$ que maximice el criterio de Fisher: $J(w) = \frac{w^T S_B w}{w^T S_W w}$.
Esto se resuelve encontrando los autovalores ($\lambda$) y autovectores ($v$) de la matriz:
$$S_W^{-1} S_B \mathbf{v} = \lambda \mathbf{v}$$

#### Paso 4: Seleccionar los nuevos ejes
Los autovectores con los autovalores más grandes son las direcciones que mejor separan las clases.
*Nota importante:* Si tienes $C$ clases, LDA puede producir como máximo $C-1$ componentes.

---

### 3. Ejercicio Práctico (Cálculo Manual Simplificado)

Imagina que tenemos 2 clases de puntos en 1D (para que sea fácil de calcular) y queremos encontrar el mejor separador.

**Datos:**
*   **Clase 1 ($C_1$):** $\{1, 2\}$
*   **Clase 2 ($C_2$):** $\{4, 5\}$

#### 1. Medias:
*   $m_1 = (1+2)/2 = \mathbf{1.5}$
*   $m_2 = (4+5)/2 = \mathbf{4.5}$
*   Media global $m = (1+2+4+5)/4 = \mathbf{3.0}$

#### 2. Dispersión Intra-clase ($S_W$):
*   Para $C_1$: $(1 - 1.5)^2 + (2 - 1.5)^2 = 0.25 + 0.25 = \mathbf{0.5}$
*   Para $C_2$: $(4 - 4.5)^2 + (5 - 4.5)^2 = 0.25 + 0.25 = \mathbf{0.5}$
*   **$S_W$** = $0.5 + 0.5 = \mathbf{1.0}$

#### 3. Dispersión Entre-clases ($S_B$):
*   $n_1(m_1 - m)^2 + n_2(m_2 - m)^2$
*   $2(1.5 - 3.0)^2 + 2(4.5 - 3.0)^2$
*   $2(-1.5)^2 + 2(1.5)^2 = 2(2.25) + 2(2.25) = \mathbf{9.0}$

#### 4. Criterio de Fisher:
El objetivo es maximizar $S_B / S_W$.
En este caso: $9.0 / 1.0 = \mathbf{9.0}$.
Si tuviéramos otra forma de proyectar los datos y el resultado fuera menor a 9, significaría que la separación es peor. LDA garantiza que esta dirección es la óptima.

---

### 4. Diferencias entre PCA y LDA

| Característica | PCA | LDA |
| :--- | :--- | :--- |
| **Tipo de aprendizaje** | No supervisado (ignora etiquetas). | Supervisado (usa etiquetas). |
| **Objetivo** | Maximizar la varianza. | Maximizar la separabilidad de clases. |
| **Uso principal** | Compresión, ruido, visualización. | Clasificación, reducción para modelos. |
| **Limitación de componentes** | Hasta el número de variables. | Máximo (Número de clases - 1). |

---

### 5. Supuestos de LDA (Para que funcione bien)
Para que LDA sea el modelo ideal, los datos deberían cumplir:
1.  **Distribución Normal:** Las variables siguen una campana de Gauss.
2.  **Misma Matriz de Covarianza:** Se asume que todas las clases tienen una dispersión similar (homocedasticidad).
3.  **Independencia:** Las muestras son independientes entre sí.

### Implementación en Python (Scikit-learn)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_iris

# Cargar datos (Iris tiene 3 clases)
data = load_iris()
X, y = data.data, data.target

# Aplicar LDA para reducir a 2 dimensiones
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X, y)

print(f"Varianza explicada por componentes: {lda.explained_variance_ratio_}")
```

**Resumen para la clase:** PCA busca "estética" (la forma de los datos), mientras que LDA busca "discriminación" (qué hace que un grupo sea distinto al otro).
````
