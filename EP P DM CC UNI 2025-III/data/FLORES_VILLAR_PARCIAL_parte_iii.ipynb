```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, recall_score
 
df_flights = pd.read_csv('FlightDelays.csv')  
# 1 
df_flights['dayweek'] = df_flights['dayweek'].astype('category')

# 2. Binning
df_flights['sched_dep_bins'] = pd.cut(df_flights['schedtime'], bins=8)


X = df_flights.drop(['delay', 'schedtime', 'deptime', 'date'], axis=1)
X = pd.get_dummies(X, drop_first=True)
 
y = df_flights['delay'].apply(lambda x: 1 if x != 'ontime' else 0)

# 3. Partición de Datos (60% para entrenar, 40% para validar)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.40, random_state=1)

# MODELADO 
# n_estimators=500 y random_state=1  
ada = AdaBoostClassifier(n_estimators=500, random_state=1)
ada.fit(X_train, y_train)

# Árbol de decisión único  
tree = DecisionTreeClassifier(random_state=1)
tree.fit(X_train, y_train)


y_pred_ada = ada.predict(X_val)
y_pred_tree = tree.predict(X_val)
# ---  PREGUNTAS ---
print(f"Accuracy AdaBoost: {accuracy_score(y_val, y_pred_ada):.4f}")
print(f"Accuracy Árbol Único: {accuracy_score(y_val, y_pred_tree):.4f}")
print(f"Recall (Vuelos retrasados) AdaBoost: {recall_score(y_val, y_pred_ada):.4f}")
print(f"Recall (Vuelos retrasados) Árbol Único: {recall_score(y_val, y_pred_tree):.4f}")
# a , la precision usando boosting adaptivo es mayor 
# b , la precision para vuelos retrasados mediante boosting es menor
# c, 
```
