```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# 1. Crear un dataset sintético (una curva con ruido)
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))  # Añadimos ruido para ver cómo reacciona CART

# 2. Configurar dos modelos CART de Regresión
# Un árbol con poca profundidad (bien generalizado)
# Un árbol con mucha profundidad (propenso a sobreajuste/ruido)
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)

regr_1.fit(X, y)
regr_2.fit(X, y)

# 3. Predecir valores para visualizar la curva aprendida
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)

# 4. Graficar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="Datos reales (ruidosos)")
plt.plot(X_test, y_1, color="cornflowerblue", label="CART: max_depth=2", linewidth=2)
plt.plot(X_test, y_2, color="yellowgreen", label="CART: max_depth=5", linewidth=2)

plt.xlabel("Dato de entrada (X)")
plt.ylabel("Valor predicho (Y)")
plt.title("CART: Regresión con Árboles de Decisión")
plt.legend()
plt.show()
```
