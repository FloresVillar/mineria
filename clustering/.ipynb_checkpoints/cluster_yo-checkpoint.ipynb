```python
# cluster basado en centroides
import numpy as np
import matplotlib.pyplot as plt

def distancia(x,u):
    d = 0.0
    for j in range(len(x)):
        d +=((x[j]-u[j])**2)
    return d

def kmeans(x,K,N=1000):
    n = len(x)
    d = len(x[0])
    centros = []   # cluster {k:0,}
    cluster = []
    for k in range(K):
        centros.append(x[k].copy())
    asignaciones = [0]*n
    for it in range(N):
        cambio = False
        for i in range(n):
            _cluster = None
            _distancia = None
            for k in range(K):
                dis = distancia(x[i],centros[k])
                if _distancia is None or dis < _distancia:
                    _distancia = dis
                    _cluster = k
            if asignaciones[i]!=_cluster:
                cambio = True
            asignaciones[i] = _cluster
        _centros = []
        conteos = []
        for j in range(K):
            _centros.append([0]*d) 
            conteos.append(0)
        for i in range(n):
            ind_cluster = asignaciones[i] # el punto i pertenece al cluster "cluster"
            conteos[ind_cluster] +=1 # conteo[cluster especifico] aumenta, es como progra dinamica
            for j in range(d): 
                _centros[ind_cluster][j]+=x[i][j] #x4 = (9, 8, 9, 8)
        for k in range(K):
            if conteos[k]>0:
                for j in range(d):
                    _centros[k][j]/=conteos[k]
        centros = []
        for k in range(K):
            centros.append(np.array(_centros[k],dtype=float)) 
        if not cambio:
            break
    for k in range(K): # dibujar
        puntos = []
        for i in range(len(x)):
            if asignaciones[i]== k:
                puntos.append(x[i])
        puntos = np.array(puntos)
        if len(puntos) > 0:
            cluster.append({'k':k,'centro':centros[k],'puntos':puntos})
    return centros,asignaciones,cluster
 
def graficar_cluster(M,cluster):
    x, y = M[:,0],M[:,1]
    plt.scatter(x,y) 
    for c in cluster: # dibujar
        puntos = c['puntos']
        centro = c['centro']
        plt.scatter(puntos[:,0],puntos[:,1])
        plt.scatter(centro[0],centro[1],marker='X',s=200)
        radio = 0.0
        for p in puntos:
            dx = p[0] - centro[0]
            dy = p[1] - centro[1]
            dist = (dx*dx + dy*dy)**0.5
            if dist > radio:
                radio = dist
        circulo = plt.Circle((centro[0], centro[1]), radio, fill=False)
        plt.gca().add_patch(circulo)
    plt.gca().set_aspect('equal')
    plt.show()
    
M, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
centros, asignaciones,cluster = kmeans(M,K=4)
print(centros)
print(asignaciones)
graficar_cluster(M,cluster) 
```

```python
# KMEANS CON SKLEARN
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

def kmeans_sklearn():
    # 1. Generar datos sintticos (4 grupos definidos)
    X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
    # 2. Configurar y entrenar el modelo K-Means
    # Definimos k=4 porque sabemos que hay 4 grupos
    kmeans = KMeans(n_clusters=4, n_init='auto')
    kmeans.fit(X)
    y_kmeans = kmeans.predict(X)
    centros = kmeans.cluster_centers_
    # 3. Visualizacin
    plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
    plt.scatter(centros[:, 0], centros[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroides')
    plt.title("Clustering K-Means (Centroides)")
    plt.legend()
    plt.show()
kmeans_sklearn()
```

```python
# JERARQUICO
import math

def dist(p,q):
    d = 0.0
    for i in range(len(p)):
        d+=((p[i]-q[i])**2)
    return math.sqrt(d)

def d_cluster(cl1,cl2):
    minima = None
    for p in cl1:
        for q in cl2:
            d = dist(p,q)
            if minima is None or d < minima:
                mimina = d
    return minima

def jerarquico_single(datos):
    clusters = []
    for punto in datos:
        clusters.append([punto])
    paso = 1
    while len(clusters)>1:
        #for c in clusters:
            #print(c)
        mejor_distancia = None
        mejor_i = None
        mejor_j = None
        for i in range(len(clusters)):
            for j in range(i+1,len(clusters)):
                d = d_cluster(clusters[i],clusters[j])
                if mejor_distancia is None or d < mejor_distancia:
                    mejor_distancia = d
                    mejor_i = i
                    mejor_j = j
        #print("\nSe fusionan:")
        #print(clusters[mejor_i], "y", clusters[mejor_j])
        #print("Distancia mnima =", round(mejor_distancia,3))
        nuevo_cluster = clusters[mejor_i] + clusters[mejor_j] ## [[1,1],[2,2]]
        if mejor_i > mejor_j:
            clusters.pop(mejor_i)
            clusters.pop(mejor_j)
        else:
            clusters.pop(mejor_j)
            clusters.pop(mejor_i)
        clusters.append(nuevo_cluster)
        paso+=1
    return clusters[0]

#datos = [[1,1],[2,2],[3,1],[6,5],[7,7],[8,6]]
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
clusters = jerarquico_single(X)
#print("\n",clusters)
```

```python
import math
import matplotlib.pyplot as plt

def dist(p,q):
    d = 0.0
    for i in range(len(p)):
        d+=((p[i]-q[i])**2)
    return math.sqrt(d)

def d_clusters(cl1,cl2):
    minima = None
    for p in cl1:
        for q in cl2:
            d = dist(p,q)
            if minima is None or d < minima:
                mimina = d
    return minima

def jerarquico_single(datos):
    clusters={i:[datos[i]] for i in range(len(datos))}
    historial=[]
    siguiente_id=len(datos)
    while len(clusters)>1:
        ids=list(clusters.keys())
        mejor_i,mejor_j=None,None
        mejor_dist=float("inf")
        for i in range(len(ids)):
            for j in range(i+1,len(ids)):
                d=d_clusters(clusters[ids[i]],clusters[ids[j]])
                if d<mejor_dist:
                    mejor_dist=d
                    mejor_i,mejor_j=ids[i],ids[j]
        historial.append((mejor_i,mejor_j,mejor_dist,siguiente_id))
        nuevo_cluster=clusters[mejor_i]+clusters[mejor_j]
        del clusters[mejor_i]
        del clusters[mejor_j]
        clusters[siguiente_id]=nuevo_cluster
        siguiente_id+=1
    return historial

def graficar_dendrograma(hist,n):
    pos={i:i for i in range(n)}
    altura={}
    plt.figure()
    for c1,c2,h,nuevo in hist:
        x1=pos[c1]
        x2=pos[c2]
        y1=altura.get(c1,0)
        y2=altura.get(c2,0)
        plt.plot([x1,x1],[y1,h])
        plt.plot([x2,x2],[y2,h])
        plt.plot([x1,x2],[h,h])
        pos[nuevo]=(x1+x2)/2
        altura[nuevo]=h
    plt.show()

#datos=[[1,1],[2,2],[3,1],[6,5],[7,7],[8,6]]
datos, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
hist=jerarquico_single(datos)
graficar_dendrograma(hist,len(datos))
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

def jerarquico_sklearn():
    # Usamos los mismos datos X del ejemplo anterior
    # 1. Crear el Dendrograma para decidir el nmero de grupos
    X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
    plt.figure(figsize=(10, 5))
    dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
    plt.title('Dendrograma (Jerrquico)')
    plt.xlabel('Puntos de datos')
    plt.ylabel('Distancia Eucldea')
    plt.show()
    # 2. Aplicar el clustering basado en el dendrograma
    hc = AgglomerativeClustering(n_clusters=4, metric='euclidean', linkage='ward')
    y_hc = hc.fit_predict(X)
    # 3. Visualizar
    plt.scatter(X[:, 0], X[:, 1], c=y_hc, s=50, cmap='rainbow')
    plt.title("Clustering Jerrquico")
    plt.show()
jerarquico_sklearn()
```

```python
# clustering basado en densidades
import math
import random
from sklearn.datasets import make_moons

def distancia(p,q):
    s=0.0
    for i in range(len(p)):
        d=p[i]-q[i]
        s+=d*d
    return math.sqrt(s)

def vecinos(datos,punto,eps):
    N=[]
    for q in datos:
        if distancia(punto,q)<=eps:
            N.append(q)
    return N

def expandir_cluster(datos,p,N,cluster,visitado,eps,minPts):
    cluster.append(p)
    i=0
    while i<len(N):
        q=N[i]
        if tuple(q) not in visitado:
            visitado.add(tuple(q))
            Nq=vecinos(datos,q,eps)
            if len(Nq)>=minPts:
                for x in Nq:
                    if x not in N:
                        N.append(x)
        if q not in cluster:
            cluster.append(q)
        i+=1

def dbscan(datos,eps,minPts):
    visitado=set()
    clusters=[]
    ruido=[]
    for p in datos:
        if tuple(p) in visitado:
            continue
        visitado.add(tuple(p))
        N=vecinos(datos,p,eps)
        if len(N)<minPts:
            ruido.append(p)
        else:
            cluster=[]
            clusters.append(cluster)
            expandir_cluster(datos,p,N,cluster,visitado,eps,minPts)
    return clusters,ruido


random.seed(0)
datos=[] 
for _ in range(30):
    datos.append([random.gauss(2,0.5),random.gauss(2,0.5)]) 
for _ in range(30):
    datos.append([random.gauss(7,0.5),random.gauss(7,0.5)]) 
for _ in range(10):
    datos.append([random.uniform(0,10),random.uniform(0,10)]) 
#clusters,ruido=dbscan(datos,eps=1.0,minPts=4)
X, _ = make_moons(n_samples=200, noise=0.05, random_state=0)
X = X.tolist()
clusters,ruido=dbscan(X,eps=0.3,minPts=5)

plt.figure()

for cluster in clusters:
    xs=[p[0] for p in cluster]
    ys=[p[1] for p in cluster]
    plt.scatter(xs,ys)

if ruido:
    xr=[p[0] for p in ruido]
    yr=[p[1] for p in ruido]
    plt.scatter(xr,yr,marker='x')

plt.title("DBSCAN desde cero")
plt.show()
```

```python
import matplotlib.pyplot as plt          
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# 1. Generar datos en forma de "lunas" (no esfricos)
X_moons, _ = make_moons(n_samples=200, noise=0.05, random_state=0)

# 2. Aplicar DBSCAN
# eps: radio de vecindad, min_samples: puntos mnimos para ser "denso"
dbscan = DBSCAN(eps=0.3, min_samples=5)
y_dbscan = dbscan.fit_predict(X_moons)

# Nota: El color morado oscuro (etiqueta -1) suele representar el ruido
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=y_dbscan, s=50, cmap='plasma')
plt.title("Clustering DBSCAN (Basado en Densidad)")
plt.show()
```
