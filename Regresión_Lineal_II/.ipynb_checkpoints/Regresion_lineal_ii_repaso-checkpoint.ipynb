```python
import numpy as np 

def obtener_costo(X,y,theta):
    return np.sum((X@theta-y)**2)/(2*len(y))

def regresion_lineal(X,y):
    return np.linalg.inv(X.T @ X) @ (X.T @ y)

def ejercicio1():
    datos = np.loadtxt('auto.txt')
    n, _ = datos.shape
    m = 1
    X = datos[:,0]
    y = datos[:,1]
    print(f"X\n{X}")
    Xs = np.ones((n,m+1))
    for i in range(n):
        Xs[i,1] = X[i]
    X = Xs
    print(f"X:\n{X}")
    theta = regresion_lineal(Xs,y)
    print(f"theta\n{theta}")
    costo = obtener_costo(Xs,y,theta)
    print(f"costo:\n{costo}")
    X_modelo = np.arange(0.5, 2.5 , 0.05) # es un X aleatorio
    n = len(X_modelo)
    Xs = np.ones((n,1 + 1))
    print(f"X_modelo\n{X_modelo}")
    for i in range(n):
        Xs[i,1] = X_modelo[i]
    print(f"X_modelo\n{Xs}")
    y_modelo = Xs@theta
    plt.scatter(X[:,1],y) # los datos originales
    plt.plot(Xs[:,1],y_modelo,'r--')
    plt.xlabel('peso')
    plt.ylabel('cosumo')
    plt.show()
ejercicio1()
```

```python
import numpy as np

def normalizacion(X): 
    m, n = X.shape
    Xf = np.zeros((m , n))
    for j in range(n):
        Xf[:,j] = (X[:,j] - X[:,j].mean())/X[:,j].std()
    return Xf

def X1(X):
    m , n = X.shape
    Xf = np.ones((m,n + 1))
    Xf[:,1:] = X
    return Xf
    
def ejercicio5():
    datos = np.loadtxt('bodyfat.txt')
    X = datos[:,:-1]
    y = datos[:,-1]
    X = normalizacion(X)
    print(f"X normalizado\n{X}")
    X = X1(X)
    print(f"X1\n{X}")
    m, n   = X.shape
    thetas = np.zeros(n,)
    print(f"theta\n{thetas}")
    thetas[7] = 10 
    grad = np.sum(((X@thetas) - y)*X[:,7])/m
    print(f"grad \n{grad}")
    #el resultado difiere al del profe
ejercicio5()
```

```python
import numpy as np
# osea la columna error z - y * la columna Xj y se suman 
# seria lo mismo que trasponer esa Xj y hacer el producto @ z- y , 
# con cada producto fila columna (suma) como elemento del vector gradiente

def gradiente(X,y,theta):
    m,n = X.shape
    grad = []
    # grad = (X.T @ (X@theta - y ))/m
    for j in range(n-1):
        g = np.sum((X@theta - y )*X[:,j])/m
        grad.append(g)
    return (X.T @ (X@theta - y ))/m
    #return grad
    
def ejercicio6():
    datos = np.loadtxt('bodyfat.txt')
    X = datos[:,:-1]
    y = datos[:,-1]
    m , n = X.shape
    Xf = X.copy()
    for j in range(n):
        Xf[:,j] = (X[:,j] - X[:,j].mean())/X[:,j].std()
    print(f"Xf\n{Xf}")
    Xff = np.ones((m,n+1))
    Xff[:,1:] = Xf
    print(f"xff\n{Xff}")
    theta = np.zeros(n,)
    theta[7] = 10
    grad = gradiente(X,y,theta)
    print(f"grad\n{grad}")
    
ejercicio6()





























    





```

```python
import numpy as np
import matplotlib.pyplot as plt

def ejercicio7():
    datos = np.loadtxt('bodyfat.txt')
    X = datos[:,:-1]
    y = datos[:,-1]
    Xf = X.copy()
    m , n = X.shape
    for j in range(n):
        Xf[:,j] = (X[:,j] - X[:,j].mean())/X[:,j].std()
    Xff = np.ones((m, n + 1))
    Xff[:,1:] = Xf
    theta = np.zeros(n,)
    theta[7] = 10
    N = 100
    alpha = 0.001
    valores = np.zeros(N)
    costes = np.zeros(N)
    for i in range(N):
        theta[7] = theta[7] - alpha*(np.sum((X@theta - y)*X[:,7]))/m
        valores[i] = theta[7]
        coste = np.sum((X@theta - y)**2)/(2*m)
        costes[i] = coste
    f, grid = plt.subplots(1,2)
    grid[0].plot(range(N),costes)
    grid[1].plot(range(N),valores)
    plt.show()
        
ejercicio7()

















```

```python


def grad_descendiente(X,y,alpha,N):
    m ,n =X.shape
    theta = np.zeros(n,)
    costes = np.zeros(N)
    for i in range(N):
        grad = (X.T@(X@theta - y))/m
        #print(f"grad\n{grad}")
        theta = theta - alpha*grad
        cost = np.sum((X@theta -y)**2)/(2*m)
        costes[i] = cost
    return theta, costes

inv = np.linalg.inv

def ejercicio8():
    datos = np.loadtxt('bodyfat.txt')
    X = datos[:,:-1]
    y = datos[:,-1]
    m, n = X.shape
    theta_gd, costes = grad_descendiente(X,y,alpha=0.000025,N=10000)
    n = np.arange(100)
    plt.plot(n,costes[:len(n)])
    plt.show()
    theta =inv(X.T@X)@(X.T@y)
    print(f"gradd\n{theta_gd}\n coste\n {np.sum((X@theta_gd -y)**2)/(2*m)}")
    print(f"theta:{theta} \n coste\n {np.sum((X@theta -y)**2)/(2*m)}")

ejercicio8()
        
```
