```python
import numpy as np 

def obtener_costo(X,y,theta):
    return np.sum((X@theta-y)**2)/(2*len(y))

def regresion_lineal(X,y):
    return np.linalg.inv(X.T @ X) @ (X.T @ y)

def ejercicio1():
    datos = np.loadtxt('auto.txt')
    n, _ = datos.shape
    m = 1
    X = datos[:,0]
    y = datos[:,1]
    print(f"X\n{X}")
    Xs = np.ones((n,m+1))
    for i in range(n):
        Xs[i,1] = X[i]
    X = Xs
    print(f"X:\n{X}")
    theta = regresion_lineal(Xs,y)
    print(f"theta\n{theta}")
    costo = obtener_costo(Xs,y,theta)
    print(f"costo:\n{costo}")
    X_modelo = np.arange(0.5, 2.5 , 0.05) # es un X aleatorio
    n = len(X_modelo)
    Xs = np.ones((n,1 + 1))
    print(f"X_modelo\n{X_modelo}")
    for i in range(n):
        Xs[i,1] = X_modelo[i]
    print(f"X_modelo\n{Xs}")
    y_modelo = Xs@theta
    plt.scatter(X[:,1],y) # los datos originales
    plt.plot(Xs[:,1],y_modelo,'r--')
    plt.xlabel('peso')
    plt.ylabel('cosumo')
    plt.show()
ejercicio1()
```

```python
# la funcion coste
import numpy as np

def ejercicio2():
    datos = np.loadtxt('auto.txt')
    Xs = datos[:,0]
    y = datos[:,1]
    n,_ = datos.shape 
    X = np.ones((n,2))
    for i in range(n):
        X[i,1] = Xs[i]
    theta_1 = 5 
    valores_0 = np.arange(-10,10,0.05)
    n= valores_0.shape
    costes_0 = np.zeros(n)
    for i, theta_0 in enumerate(valores_0):
        costes_0[i] = obtener_costo(X,y,[theta_0,theta_1])
    theta_0 = 0
    valores_1 = np.arange(4,14,0.05)
    n = valores_1.shape
    costes_1 = np.zeros(n)
    for i,theta_1 in enumerate(valores_1):
        costes_1[i] = obtener_costo(X,y,[theta_0,theta_1])
    f, grid = plt.subplots(1,2)
    grid[0].plot(valores_0,costes_0,'g')
    grid[1].plot(valores_1,costes_1,'b')
    plt.show()

ejercicio2()







```
