```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score

def evaluar_modelo(X,y_real,theta): # uso esta como base, se implemento en regresion logistica
    z = X @ theta
    probs = sigmoide(z)
    y_pred = (probs > 0.5).astype(int) #probs >0.5 ? y_pred = 1 : y_pred= 0
    tp = np.sum((y_real == 1)&(y_pred == 1)) # verdaderos positivos
    tn = np.sum((y_real == 0)&(y_pred == 0)) # verdaderos negativos
    fp = np.sum((y_real == 0)&(y_pred == 1)) # falsos positivos
    fn = np.sum((y_real == 1)&(y_pred == 0)) # falsos negativos
    accuracy = (tp + tn)/len(y_real)
    precision = tp/(tp + fp) if (tp + fp) > 0 else 0
    recall = tp/(tp + fn) if (tp + fn) > 0 else 0
    f1 = 2*(precision * recall)/(precision + recall) if (precision + recall) > 0 else 0
    return {"matriz":[[tn,fp],[fn,tp]],
            "accuracy":accuracy,
            "precision":precision,
            "recall":recall,
            "f1":f1}

def my_accuracy(y_real, y_pred):
    """
    Exactitud (accuracy) =  
    (TP + TN) / (TP + TN + FP + FN)
    """
    # TODO
    TP = np.sum((y_real == 1) & (y_pred == 1))
    TN = np.sum((y_real == 0) & ( y_pred==0))
    n = len(y_real)
    # En el caso de un arreglo vacío
    if n == 0:
        return 0.0 
    return (TP + TN) / n

def my_precision(y_real, y_pred):
    """
    Precisión = 
    TP / (TP + FP)
    """
   # TODO
    TP = np.sum((y_real == 1)&(y_pred == 1)) 
    FP = np.sum((y_real == 0)&(y_pred == 1))
    #precision = tp/(tp + fp) if (tp + fp) > 0 else 0
    # En caso de una división por cero (no hay positivos predichos)
    if (TP + FP) == 0:
        return 0.0  
    return TP / (TP + FP)

def my_recall(y_real, y_pred):
    """
    Exhaustividad (recall) = 
    TP / (TP + FN)
    """
    TP = np.sum((y_real == 1) & (y_pred == 1))
    FN = np.sum((y_real == 1)&(y_pred == 0))
    # TODO
    #recall = tp/(tp + fn) if (tp + fn) > 0 else 0
    # En caso de una división por cero (no hay positivos verdaderos)
    if (TP + FN) == 0:
        return 0.0   
    return TP / (TP + FN)

def my_f1(y_real, y_pred):
    """
    F1-score =
    2 * (Precision * Recall) / (Precision + Recall)
    """
    # TODO
    #f1 = 2*(precision * recall)/(precision + recall) if (precision + recall) > 0 else 0
    precision = my_precision(y_real,y_pred)
    recall = my_recall(y_real,y_pred)
    denominador = precision + recall
    # En caso de división por cero
    if denominador == 0:
        return 0.0 
    return 2 * (precision * recall) / denominador

# Verificación
y_true_c = np.array([0,1,1,0,1,0,0,1])
y_pred_c = np.array([0,1,0,0,1,0,1,1])

# Nos aseguramos que las funciones de sklearn.metrics usan zero_division=0 para coincidir con nuestro comportamiento
# cuando el denominador es cero.
assert np.isclose(my_accuracy(y_true_c, y_pred_c), accuracy_score(y_true_c, y_pred_c))
assert np.isclose(my_precision(y_true_c, y_pred_c), precision_score(y_true_c, y_pred_c, zero_division=0))
assert np.isclose(my_recall(y_true_c, y_pred_c), recall_score(y_true_c, y_pred_c, zero_division=0))
assert np.isclose(my_f1(y_true_c, y_pred_c), f1_score(y_true_c, y_pred_c, zero_division=0))
print("Verificaciones from-scratch OK")
```
