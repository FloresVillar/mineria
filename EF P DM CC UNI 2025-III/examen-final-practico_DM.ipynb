### Ejercicio 1: Optimización de Variedades y Reconstrucción Lineal en Reconocimiento Facial (10 puntos)

**Contexto del Proyecto:**  
Usted ha sido contratado como Ingeniero de Minería de datos para una firma de ciberseguridad. Debido a la alta dimensionalidad de las imágenes (4096 píxeles), el sistema actual es ineficiente. Su misión es diseñar un pipeline de preprocesamiento que optimice la compresión de datos mediante **PCA** y la visualización de clusters mediante **t-SNE**, garantizando la integridad topológica de la información.

3.  Grafique la varianza acumulada, la recta de referencia y marque claramente el punto del codo hallado.

### Parte 2: Optimización de la Proyección No Lineal (t-SNE) (2 puntos)
